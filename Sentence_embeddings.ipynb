```python
from transformers.utils import logging
```

```python
logging.set_verbosity_error()
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model=SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sent1=['Cat is in the hallway', 'I play the guitar', 'The movie is good']
```

```python
embed1=model.encode(sent1,convert_to_tensor=True)
```

```python
embed1
```

```python
sent2=['The dog is on the couch', 'I watch TV', 'The like the movie']
```

```python
embed2=model.encode(sent2,convert_to_tensor=True)
```

```python
embed2
```

```python
from sentence_transformers import util
```

```python
cosine_scores = util.cos_sim(embed1,embed2)
```

```python
print(cosine_scores)
```

```python
for i in range(len(sent1)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(sent1[i],
                                                 sent2[i],
                                                 cosine_scores[i][i]))
```
